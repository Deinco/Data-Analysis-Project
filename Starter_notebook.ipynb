#### Volume of payments per payment channel
Earlier, the number of payments per payment channel was analysed. It may also be useful to determine the total volume of payments obtained via each payment channel as lesser number of transactions may not always mean lesser total amount.

```python
payment_channel_volume = sns.catplot(data = supermarkets, kind = 'bar', x = 'Payment', 
                                     y = 'Total', estimator = sum, legend = False)
payment_channel_volume.fig.suptitle("Payments Channel Volume")
```

Whereas epay was the most used payment channel, cash payments had a greater total volume. This was followed by epay and then card payments.
